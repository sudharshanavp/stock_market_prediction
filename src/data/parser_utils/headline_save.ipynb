```python
save_path = "../../../data/raw/news/"
```

```python
for index, row in out_df.iterrows():
    
    if os.path.isfile("../../../data/raw/news/" + row["Symbol"] + ".csv"):
        continue
    
    try:
        tag = row["tags"]
        page_links = money_control.get_page_links(tag)
        article_list = money_control.get_article_list(page_links)
        headlines = money_control.get_headlines(article_list)
        article_links = money_control.get_article_links(article_list)

        news_df = pd.DataFrame()
        news_df["Headlines"] = pd.Series(headlines)
        news_df["Links"] = pd.Series(article_links)

        news_df.to_csv(save_path + row["Symbol"] + ".csv")
    except:
        print("Exception " + row["Symbol"])
```
